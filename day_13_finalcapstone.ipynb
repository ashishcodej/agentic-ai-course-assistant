```python
# CELL 5: Create Embeddings and ChromaDB

from sentence_transformers import SentenceTransformer
import chromadb

# Load embedding model
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Prepare data
texts = [doc["text"] for doc in documents]
ids = [doc["id"] for doc in documents]
metadatas = [{"topic": doc["topic"]} for doc in documents]

# Create embeddings
embeddings = embedder.encode(texts).tolist()

# Create ChromaDB client
client = chromadb.Client()

# Create collection
collection = client.create_collection(name="course_assistant")

# Add data to DB
collection.add(
    documents=texts,
    embeddings=embeddings,
    ids=ids,
    metadatas=metadatas
)

print("ChromaDB created successfully ✅")
```

```python
# CELL 6: Test Retrieval

query = "What is LangGraph?"

# Convert query to embedding
query_embedding = embedder.encode([query]).tolist()

# Search in ChromaDB
results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

# Print results
for i, doc in enumerate(results["documents"][0]):
    print(f"\nResult {i+1}:")
    print(doc)
```

```python
# CELL 7: Define State (Agent Brain)

from typing import TypedDict, List

class CapstoneState(TypedDict):
    
    # User input
    question: str
    
    # Conversation memory
    messages: List[dict]
    
    # Routing decision (retrieve / tool / memory)
    route: str
    
    # Retrieved context from ChromaDB
    retrieved: str
    
    # Source documents
    sources: List[str]
    
    # Tool output (if any tool is used)
    tool_result: str
    
    # Final answer generated by LLM
    answer: str
    
    # Evaluation score (RAGAS / faithfulness)
    faithfulness: float
    
    # Retry counter for evaluation loop
    eval_retries: int

print("State defined successfully ✅")
```

```python
# CELL 8: Define Nodes

from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

pipe = pipeline("text-generation", model="distilgpt2")

llm = HuggingFacePipeline(pipeline=pipe)


# 1. MEMORY NODE
def memory_node(state):
    if "messages" not in state or state["messages"] is None:
        state["messages"] = []
        
    state["messages"].append({
        "role": "user",
        "content": state["question"]
    })
    
    return state


# 2. ROUTER NODE
def router_node(state):
    question = state["question"].lower()
    
    if "calculate" in question or "math" in question:
        state["route"] = "tool"
    else:
        state["route"] = "retrieve"
    
    return state


# 3. RETRIEVAL NODE
def retrieval_node(state):
    query = state["question"]
    
    query_embedding = embedder.encode([query]).tolist()
    
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=2
    )
    
    retrieved_docs = results["documents"][0]
    
    state["retrieved"] = "\n".join(retrieved_docs)
    state["sources"] = results.get("metadatas", [{}])[0]
    
    return state


# 4. TOOL NODE (simple calculator)
def tool_node(state):
    question = state["question"]
    
    try:
        result = eval(question)
        state["tool_result"] = str(result)
    except:
        state["tool_result"] = "Unable to calculate"
    
    return state


# 5. ANSWER NODE
def answer_node(state):
    context = state.get("retrieved", "")
    question = state["question"]

    prompt = f"""
Context:
{context}

Question:
{question}

Answer:
"""

    response = llm.invoke(prompt)

    # Clean output (important)
    answer = response.split("Answer:")[-1].strip()

    state["answer"] = answer

    return state


print("All nodes created successfully ✅")
```

```python
# CELL 9: Build LangGraph Workflow

from langgraph.graph import StateGraph

# Create graph
builder = StateGraph(CapstoneState)

# Add nodes
builder.add_node("memory", memory_node)
builder.add_node("router", router_node)
builder.add_node("retrieve", retrieval_node)
builder.add_node("tool", tool_node)
builder.add_node("answer", answer_node)

# Set entry point
builder.set_entry_point("memory")

# Define flow
builder.add_edge("memory", "router")

# Conditional routing
builder.add_conditional_edges(
    "router",
    lambda state: state["route"],
    {
        "retrieve": "retrieve",
        "tool": "tool"
    }
)

# Continue flow
builder.add_edge("retrieve", "answer")
builder.add_edge("tool", "answer")

# Compile graph
graph = builder.compile()

print("LangGraph agent created successfully ✅")
```

```python
# CELL 11: Testing (10 Questions + PASS/FAIL)

test_questions = [
    "What is Agentic AI?",
    "Explain ReAct framework",
    "What is memory in agents?",
    "What are embeddings?",
    "What is ChromaDB?",
    "What is LangChain?",
    "What is LangGraph?",
    "What is RAG?",
    "What is RAGAS?",
    "What happens in Day 13?"
]

results_list = []

for q in test_questions:
    result = graph.invoke({
        "question": q,
        "messages": [],
        "eval_retries": 0
    })
    
    answer = result["answer"]
    
    # Simple PASS/FAIL logic
    if len(answer.strip()) > 20:
        status = "PASS"
    else:
        status = "FAIL"
    
    results_list.append((q, answer, status))

# Print results
for i, (q, ans, status) in enumerate(results_list, 1):
    print(f"\nTest {i}: {status}")
    print("Q:", q)
    print("A:", ans)
```
